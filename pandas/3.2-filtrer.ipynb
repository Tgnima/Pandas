### Filtrage avec plusieurs conditions

Sélectionnez les hommes de seconde classe qui sont morts qui ont entre 20 et 30 ans.

```python
hommes = (df.sex == "male")
scd_class = (df["class"] == "Second")
age_20_30 = (df.age >= 20) & (df.age < 30)
```

```python
df[hommes & scd_class & age_20_30]
```

```python
passager_young_middle_class = hommes & scd_class & age_20_30
```

```python
df[passager_young_middle_class].shape
```

#### Opérateurs booléen

```python
# And : Qui ont entre 20 et 30 ans et sont en seconde classe
df[age_20_30 & scd_class]
df[~age_20_30]  # Not : Qui n'ont pas entre 20 et 30 ans
# Or : Qui ont entre 20 et 30 ans, ou sont en seconde classe
df[age_20_30 | scd_class]
# Xor : Qui ont soit entre 20 et 30 ans mais pas en seconde classe, soit en 2nd classe mais pas entre 20 et 30
df[age_20_30 | scd_class]
```

Sélectionnez les hommes qui sont soit en seconde classe, soit qui ont entre 20 et 30 ans.

```python
df[hommes & (scd_class | age_20_30)]
```

Sélectionnez les hommes qui sont soit en seconde classe, soit qui ont entre 20 et 30 ans, mais qui ne sont pas de seconde classe et qui ont entre 20 et 30 ans.

```python
df[hommes & (scd_class ^ age_20_30)]
```
