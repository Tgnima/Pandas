#### Selection par index pour par position

Deux méthodes (canonique) pour sélectionner des sous ensemble de dataframe. 
`loc` et `iloc`.

- `loc` permet de sélectionner par l'`index` pandas, en d'autre terme en fonction du "nom" de la ligne ou du nom de la colonne.
- `iloc` permet de sélectionner par la *position*  (le `i` de `iloc` vient de ''integer position based'' )

```python
df_disney = pd.DataFrame(
    [["Riri","Canard", 1937, .80],   
     ["Dingo", "Chien", 1932, 1.8],#### Ceci est un dictionnaire de dictionnaire
    ["Mickey",  "Souris", 1928, 1],
     ["Donald", "canard", np.nan, 1.05]
   ], columns=["nom", "Espece", "date_creation", "taille"]).set_index("nom")
```

```python
df_disney
```

```python
df_disney.loc["Riri",:]  #Peut être raccourcie en .loc["Riri"]
```

```python
df_disney.iloc[0,:]
```

Donc en utilisant loc et iloc, on slice la dataframe, le premier élément correspondant à la ligne, et le deuxième à la colonne.

`.loc["start_row_name":"end_row_name", "start_column_name":"end_column_name]`

```python
df_disney.loc["Riri"] #Selectionne la ligne de Riri
df_disney.loc["Riri":"Mickey"] #Sélectionne les ligne de Riri à Mickey inclu...
df_disney.loc["Riri":"Mickey", "Espece":"date_creation" ] # Sélectionne les lignes de Riri à Michey et de Espece à date_creation
df_disney.loc[["Riri", "Donald", "Dingo"], ["taille", "Espece"]]
```
