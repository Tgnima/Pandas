#### Plusieurs conditions

```python
df[(df.Ville ==  "MARSEILLE") & ~ df["Prix E10"].isna() ].sort_values(by="Prix E10",ascending=True).head(10)
```

```python
df[(df.Ville ==  "MARSEILLE")  
  & (~ df["Prix E10"].isna()) 
  & (df.Services.str.contains("gonflage", case=False))
  & (df.Services.str.contains("restau", case=False, na=False))
  ].sort_values(by="Prix E10",ascending=True).head(10)
```

```python
df[(df.Ville ==  "MARSEILLE")  
  & (~ df["Prix Gazole"].isna()) 
  ].sort_values(by="Prix Gazole",ascending=True).head(20)
```
